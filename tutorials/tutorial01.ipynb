## Setting up an MNIST Corpus

<a href="#First-steps-towards-using-the-lrn2-framework">top</a>

The <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> class is the central input data processor and supplier module of the lrn2 framework. The <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> organizes data import from files or other sources, as well as data conversion, in order to make it suitable as input to Neural Networks (NNs). In addition, the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> is able to organize sequential data as a set of subsequences of desired length (ngram representation).

Data import and data conversion is executed by external classes, which can be specified for different file formats, domains, and input representations.

In the example below, we set up a <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a>, to load and organize the well-known <a href="http://yann.lecun.com/exdb/mnist/">MNIST handwritten digits database.</a>

#### Import necessary modules

```python
import os
from functools import partial
from lrn2.data.formats.mnist import load_mnist_files, MNISTInterface
from lrn2.data.domain.image import ImageBinaryVP
from lrn2.data.corpus import Corpus
```

#### Specifying the MNIST Viewpoint

Viewpoints are a flexible concept of the lrn2 framework which allow for a reusable format-specific loading of files and domain-specific conversion of data. A <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.formats.html">format-specific Interface</a> knows how data should be read from a certain filetype and provides interfaces from which <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html">domain-specific Viewpoints</a> may collect the data in a standardized format and convert it into a representation suited as input for NNs (e.g. binary representation), or defines how the data should be plotted.

In the following, we define a new class <code>MnistVP</code> as a combination (mix-in) of the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.formats.html#module-lrn2.data.formats.mnist">MNISTInterface</a> (format-specific interface), and the <a href = "http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html#module-lrn2.data.domain.image">ImageBinaryVP</a> (domain-specific viewpoint). Note that as the interface implements abstract methods of the viewpoint, it has to be the first parameter in the class specification.

```python
class MnistVP(MNISTInterface, ImageBinaryVP): pass
```

<a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.formats.html#module-lrn2.data.formats.mnist">lrn2.data.formats.mnist</a> implements methods for loading the actual pictures of handwritten digits from the MNIST archive and with <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.formats.html#module-lrn2.data.formats.mnist">MNISTInterface</a> it provides a standardized interface for collecting the pictures.
The <a href = "http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html#module-lrn2.data.domain.image">ImageBinaryVP</a> may then consult this interface to receive a standardized image representation. From there, it provides methods for conversion to a NN input representation or for plotting those images. Note that the <a href = "http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html#module-lrn2.data.domain.image">ImageBinaryVP</a> can process any binary images, independent of which source they were loaded from.

#### Instantiating the derived Viewpoint

To instantiate the <a href = "http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html#module-lrn2.data.domain.image">ImageBinaryVP</a>, we need to pass the shape of the images it will process, while the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.formats.html#module-lrn2.data.formats.mnist">MNISTInterface</a> does not expect any parameters to its init method. Therefore, also the derived viewpoint <code>MnistVP</code> needs to know just the shape of the MNIST images, which is <code>(28,28)</code>.

```python
viewpoints = (MnistVP((28,28)),)
```

#### Preparing the file loader

As the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> class expects a standardized method signature for the file loader with only one parameter pointing to the source of the data, we need to pre-specify the second parameter of <code>load_mnist_files</code> via the <code>partial</code> function. This function returns another function, which calls <code>load_mnist_files</code> as it would have <code>subsets = ('train',)</code> as default argument.

```python
file_loader = partial(load_mnist_files, subsets = ('train',))
```

#### Building the Corpus

Finally, we can instantiate the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> class and we pass the <code>file_loader</code> and the one-element list of viewpoints.

```python
corpus = Corpus(file_loader, viewpoints, sparse_data = False)
```

Subsequently, we call the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#lrn2.data.corpus.Corpus.load_files">load_files()</a> method where we pass a reference to the data source. This causes the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> to consult the file loader of the format specific viewpoint. Note that <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#lrn2.data.corpus.Corpus.load_files">load_files()</a> can be called repeatedly with different data source references to incrementally fill the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a>.

```python
data = os.path.join(os.getcwd(), 'data', 'mnist', 'mnist.pkl.gz')
corpus.load_files(data)
```

By calling the method <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#lrn2.data.corpus.Corpus.set_to_ngram">set_to_ngram()</a> with <code>ngram_size = 1</code>, we define that our data is not sequential and each ngram consists of a separate image. Note that <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#lrn2.data.corpus.Corpus.set_to_ngram">set_to_ngram()</a> does the actual conversion of the raw data imported by the file loader to an input representation suited for NNs by consulting the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.domain.html">domain-specific viewpoints</a>. The method <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#lrn2.data.corpus.Corpus.shuffle_instances">shuffle_instances()</a> randomizes the order of our ngrams, which ensures that there is no correlation of the data and its order.

```python
# Create training instances in NN input representation
corpus.set_to_ngram(ngram_size = 1)
# Randomize ngrams
corpus.shuffle_instances()

print "images in corpus = {0}".format(corpus.n_ngrams)
print "size of an image = {0}".format(corpus.n_dimensions)
```

## The Configuration File

<a href="#First-steps-towards-using-the-lrn2-framework">top</a>

Specifying the architecture and the training parameters of an NN stack (e.g. which and how many hidden units, which training procedure, etc.) is done with a configuration file. This file may consist of multiple nested sections, as shown in the following example (execute the code snippet below).

```python
# Plot the configuration file
with open("config_model.ini", 'r') as config_file:
    print config_file.read()
```

This configuration file specifies a two-layered RBM stack with an RBM using persistent contrastive divergence (Tieleman, 2008) and weight regularization based on (H. Goh, 2010) in the first layer and second layer, and a simple NN layer with linear hidden units and softmax activation.

There is a distinct parameter section for training the whole STACK with similar parameters than those of single layers. However, this is just for giving an idea of what is possible - in this demo we won't use the STACK section. The config file is a copy of that in the <code>examples/mnist_pretrain</code> demo, please refer to that demo to see how to use the configuration file.

With that configuration file in hand, we can now start to build and train the RBM stack.

## Creation and training of an RBM stack

<a href="#First-steps-towards-using-the-lrn2-framework">top</a>

#### Import necessary modules

```python
import pylab
from lrn2.util.config import get_config
from lrn2.nn_bricks.make import make_net
from lrn2.nn_bricks.train import train_layer_wise

# Magic command for IPython Notebook to plot in browser
%matplotlib inline
```

#### Plot filters (i.e. weights, receptive fields) of the hidden units:

Creation and training of our RBM stack is a one-liner each. Before, we have to load the <a href="#The-Configuration-File">configuration file</a> as a dictionary using <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.util.html?highlight=get_config#lrn2.util.config.get_config">get_config()</a> which we then use as an argument to the <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.nn_bricks.html#lrn2.nn_bricks.train.train_layer_wise">train_layer_wise()</a> method. We also have to decide for a <code>run_keyword</code>, which determines the name of the folder where plots and states of the RBM will be written to during training. In addition, we bind the training data contained in our <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.data.html#module-lrn2.data.corpus">Corpus</a> to the 'input' variable using a dictionary.

The parameter <code>re_train</code> determines, if we want to ignore saved states of the RBMs during previous trainings. Note that the parameter <code>dump_interval</code> in our configuration file specifies that every <code>N</code> epochs of training, the weights and biases of the RBM are getting saved in a **backup** file. If a training is finished, or interrupted by [strg+c], a **cache** file will be written. <a href="http://lrn2cre8.ofai.at/lrn2/doc/lrn2.nn_bricks.html#lrn2.nn_bricks.train.train_layer_wise">train_layer_wise()</a> searches for this file. If found, it is used to reconstruct an RBM and the training of the next layer begins. The parameter <code>re_train</code> overrides that mechanism and trains the whole stack anew. If you want to load a specific **backup** file, overwrite the last **cache** file with it.

```python
# Open configuration file (as dictionary)
config = get_config("config_model.ini")
# Create a list of RBM layers, according to the configuration file
rbm_layers = make_net(config, corpus.size)
# Build and train RBM stack according to configuration file
trained_rbm_layers = train_layer_wise(net = rbm_layers,
                                      data = {'input': corpus.ngram_data},
                                      config = config,
                                      run_keyword='tutorial',
                                      tile_fun = corpus.tile_ngram_data,
                                      re_train = True)
```
